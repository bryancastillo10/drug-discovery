### ChemBL Database

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#Query the target protein from the ChemBL database
target = new_client.target
target_query = target.search('acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets.head()
```

```python
targets.shape
```

```python
#Filter to the 5th entry (index 4)
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type="IC50")
df = pd.DataFrame.from_dict(res)
df
```

```python
# Saving the raw data as csv file
df.to_csv('bioactivity_data.csv', index=False)
```

```python
# Handle missing data
df2 = df[df.standard_value.notna()]
# Also filtered with non-null canonical smiles (string data type representation of a chemical structure)
df2 = df2[df.canonical_smiles.notna()]
df2
```

### Data pre-processing

```python
# Drop data duplicates on canonical smiles
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
# Select only this three columns
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

```python
# Save as csv the selected data
df3.to_csv('02_bioactivity_data_preprocessed.csv', index=False)
```
